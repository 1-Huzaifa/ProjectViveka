```python

from docx import Document

def extract_paragraphs_from_docx(docx_path):

    doc = Document(docx_path)
    return [para.text for para in doc.paragraphs if para.text.strip() != ""]

def tokenize_paragraph(paragraph, block_size=512):
    """input: (paragraph,block_size=512), output : block size of length 512 characters """
    blocks = []
    start = 0
    end = block_size

    while start < len(paragraph):
        block = paragraph[start:end]
        while block and block[-1] not in [".", "!", "?", "\n",] and end < len(paragraph):

            end -= 1
            block = paragraph[start:end]


        if not block:
            end = start + block_size
            block = paragraph[start:end]

        blocks.append(block.strip())
        start = end
        end = start + block_size
    return blocks

def save_text_to_docx(text, docx_file):
    try:
        doc = Document()
        doc.add_paragraph(text)
        doc.save(docx_file)
        return "Text saved to Word document successfully."
    except Exception as e:
        print("exception 2")
        return str(e)


docx_path = "1. Raw - Chapter 12 -Bhagavad Gita - Theme 1 (1) (1).docx"

paragraphs = extract_paragraphs_from_docx(docx_path)

tokenized_blocks = []
for paragraph in paragraphs:
    tokenized_blocks.extend(tokenize_paragraph(paragraph))


for idx, block in enumerate(tokenized_blocks):
    print(f"Block {idx + 1}:\n{block}")
    docx_file = f"/content/block{idx+1}.docx"
    save_text_to_docx(block, docx_file)
```
